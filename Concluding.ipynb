```python
import os
import sys
import pandas as pd
filepath = os.path.dirname(os.path.abspath("")) + "/ToPython"
```

```python
dados = ['/dados','/dadosEntropy']
dados = ['/dados']

data_types = ['poly','ts']
atributes = ['bytes','pacotes']

for dado in dados:
    for data_type in data_types:
        for atribute in atributes:
            fullpath = filepath + dado + '/erro_' + data_type + '_'+atribute+'.csv'
            print(fullpath)
            display(pd.read_csv(fullpath))
```
